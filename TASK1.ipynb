```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%y')
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month_name()
data['Day'] = data['Date'].dt.day
data['Weekday'] = data['Date'].dt.day_name()
```

```python
data.head()
```

```python
Order_status = data['Status'].value_counts()
print(Order_status)
```

```python
Total_revenue_by_Month = data.groupby('Month')['Total Sales'].sum().sort_values(ascending=False)
print(Total_revenue_by_Month)
```

```python
Total_revenue_by_Category = data.groupby('Category')['Total Sales'].sum().sort_values(ascending=False)
print(Total_revenue_by_Category)
```

```python
Total_revenue_by_Product = data.groupby('Product')['Total Sales'].sum().sort_values(ascending=False).head(5)
print(Total_revenue_by_Product)
```

```python
Total_revenue_by_Location = data.groupby('Customer Location')['Total Sales'].sum().sort_values(ascending=False).head(3)
print(Total_revenue_by_Location)
```

```python
Total_revenue_by_Payment_method = data['Payment Method'].value_counts()
print(Total_revenue_by_Payment_method)
```

```python
Week_days_sales_distribution = data['Weekday'].value_counts()
print(Week_days_sales_distribution)
```
